```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
df = pd.read_csv('/content/StudentsPerformance.csv')
df.head()

#Data cleaning
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df.head()
print("Rows and columns:", df.shape)
print("\nColumn Data Types:")
print(df.dtypes)
df.describe()

#Data analysis and visualization
df['average_score'] = df[['math_score', 'reading_score', 'writing_score']].mean(axis=1)
df[['math_score', 'reading_score', 'writing_score', 'average_score']].head()


sns.barplot(x='gender', y='math_score', data=df)
plt.title('Math Score by Gender')
plt.show()

sns.barplot(x='lunch', y='average_score', data=df)
plt.title('Average Score by Lunch Type')
plt.ylabel('Average Score')
plt.xlabel('Lunch Type')
plt.show()

sns.scatterplot(x='reading_score', y='writing_score', hue='gender', data=df)
plt.title("Reading vs Writing Scores by Gender")
plt.show()

numeric_df = df.select_dtypes(include='number')
corr = numeric_df.corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix (Numeric Columns Only)")
plt.show()

```

###  Key Observations from Student Performance Analysis

---

####  1. Dataset Overview
- 1000 student records
- Includes scores in math, reading, and writing, and demographic factors

####  2. Data Cleaning
- Cleaned column names
- Created `average_score` column

####  3. Score Summary
- Reading and writing scores tend to be higher than math
- Average scores give a better overall performance indicator

####  4. Math Score by Gender
- Males slightly outperform females in math
- Both genders have wide score ranges

####  5. Average Score by Lunch Type
- Standard lunch students scored higher on average
- Suggests socioeconomic influence on performance

####  6. Correlation Matrix
- Strong relationship between reading, writing, and average scores
- Reading and writing are highly correlated (0.95)
- All three subjects contribute strongly to academic performance
